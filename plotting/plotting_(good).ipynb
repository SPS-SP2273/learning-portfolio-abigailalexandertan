## 1 Comparing the two ‘dialects’

### 1.1 Some nomenclature

```python
# matplotlib figure = canvas to draw stuff on 
# matplotlib axis = individual mathematical axis. 
                    #can have multiple acis!
```

### 1.2 We need some data.

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### 1.3 Here comes the comparison

```python
# pyplot interface 

from matplotlib import pyplot as plt

plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
plt.show()
```

```python
# OO interface

from matplotlib import pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
plt.show()
```

##### Notes:

- Both Pyplot and OO interface produces the same plot.
- Use pyplot for quick plots.
- Use OO for complex plots. 
- OO interface need to start by using subplots(). This will ask matplotlib to create a figure and an axis. 


## 2 Using the OO Interface

```python
fig, ax = plt.subplots(ncols=1, nrows=2)

# This gives 2 axes so that we can plot in both panes.

# ncols=1 --> to ask for one column
# nrows=2 --> to ask for two rows
```

### 2.3 Plots in a column

```python
# To plot the first plot:

fig, ax = plt.subplots(ncols=1, nrows=2)

ax[0].plot(x, cos_x, label='cos x')
```

```python
# To plot the second plot:

fig, ax = plt.subplots(ncols=1, nrows=2)

ax[1].plot(x, sin_x, label='sin x')
```

### 2.4 Legends

```python
# In order to show legends, we need to call it for each axis.

# To add legend for first plot:

fig, ax = plt.subplots(ncols=1, nrows=2)
ax[0].plot(x, cos_x, label='cos x')
ax[0].legend()
```

```python
# To add legend for second plot:

fig, ax = plt.subplots(ncols=1, nrows=2)
ax[1].plot(x, sin_x, label='sin x')
ax[1].legend()
```

```python
# Instead of doing individually, we can ask to loop it 

fig, ax = plt.subplots(ncols=1, nrows=2)
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

for a in ax:
    a.legend()
    
```

```python
# In order to add grid to each of the axes:

fig, ax = plt.subplots(ncols=1, nrows=2)
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

for a in ax:
    a.legend()
    a.grid(alpha=.25)
```

### 2.5 Tweaks

```python
# Change the size of the plot by specifying figure size (figsize).
# figsize=(5, 10) means size is 10 x 5 inches
# Ask that the plots share the x axis using (sharex).

fig, ax = plt.subplots(nrows=2, ncols=1,
                     figsize=(5, 10),
                     sharex = True)

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0) 
    
# To add x label to OO interface
ax[1].set_xlabel('$x$')

# To fill the plot between 0 and the plot values:
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].fill_between(x, 0, sin_x, alpha=.25)

# To add a super title to the figure:
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

# To ask matplotlib to make adjustments to make the plot look nice
fig.tight_layout()

plt.show()
```

## 3 More rows and columns

### 3.1 Using ax

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10, 4),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readability:
# ax.flatten() takes the 2D array and flattens it into 1D array.
# The unpacking takes care of the anssignments.
top_left, top_right, bottom_left, bottom_right = ax.flatten()

# x data for plotting:
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(10, 20, 100)

# To access each of the axes:
top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

for a in ax.flatten():
    a.grid(alpha=.25)
    
plt.tight_layout()
plt.show()
```

### 3.2 Sharing axes

```python
# Plots can be compacted by sharing x and y axes 
# sharex and sharey

fig, ax = plt.subplots(nrows=2, ncols=2,
                      figsize=(10, 4),
                       sharex=True, sharey='row')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

x1 = np.linspace(0, 10, 100)
x2 = np.linspace(10, 20, 100)


top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

for a in ax.flatten():
    a.grid(alpha=.25)
    
plt.tight_layout()
plt.show()
```

### 3.3 Accessing all axes

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                      figsize=(10, 4),
                       sharex=True, sharey='row')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

x1 = np.linspace(0, 10, 100)
x2 = np.linspace(10, 20, 100)


top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

# The longer way of changing all the axes:
top_left.grid(alpha=.25)
top_right.grid(alpha=.25)
bottom_left.grid(alpha=.25)
bottom_right.grid(alpha=.25)
    
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                      figsize=(10, 4),
                       sharex=True, sharey='row')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

x1 = np.linspace(0, 10, 100)
x2 = np.linspace(10, 20, 100)


top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

# The easier, shorter way of changing all the axes with a for loop:
for a in ax.flatten():
    a.grid(alpha=.25)
    
plt.tight_layout()
plt.show()
```

## 4 Other useful plots

### 4.1 Histograms

```python
data = {}

filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
data['All'] = np.loadtxt(filename, skiprows=1)

name = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
data['NUS'] = np.loadtxt(filename, skiprows=1)

plt.style.use('bmh')

plt.hist([data['All'], data['NUS']],
         bins=50,
         label=['All', 'NUS']
        )

plt.xlabel('Mean of Basic Monthly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

### 4.2 Scatter plots

```python
# Are created by putting a marker at a specified (x, y) point.
# We need values for x so use range() and len() to generate a list.

data = {}
for label in ['All', 'NUS']:
    filename = f'sg-gov-graduate-employment-survey_basic_monthly_mean_{label.lower()}.csv'
    data[label] = np.loadtxt(filename, skiprows=1)

plt.style.use('seaborn-darkgrid')

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Monthly Earning (S$)')
plt.legend()
plt.show()
```

### 4.3 Bar charts

```python
# barh() --> to make horizontal bars

student_numbers = {'Life Sciences': 14,
                  'Physics': 12,
                  'Chemistry': 8,
                  'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers)
plt.xlabel('Majors')
plt.ylabel('Number of Students')
plt.show()
```

### 4.4 Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}

majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers,
       labels=majors,
       autopct='%1.1f%%',     # how to format the percentages
       startangle=-90
       )

plt.title('Percentage of each major')
plt.show()
```

```python
#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,
                         figsize=(12, 8),
                         sharex= True, sharey='row')

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='red', label='$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='red',
                linestyle=':', label='$\cos^2 x$')
axes[0, 0].set_title('$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -1, color='red', alpha=.125)
axes[0, 0].legend(loc='upper right', frameon=True)
#axes[0, 0].text(0, 0, 'I am [0, 0]!', fontsize=30,
                #horizontalalignment='center')

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='green', label='$\sin x$')
axes[0, 1].fill_between(x, sin_x, -2, color='green', alpha=.25)
axes[0, 1].set_title('$\sin x$ & $\sin^2x$')
axes[0, 1].plot(x, sin_x**2, label='$\sin^2 x$')
#axes[0, 1].set_xlabel('Angle (radians)')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].legend(loc='upper right', frameon=True)
#axes[0, 1].text(0, 0, 'I am [0, 1]!', fontsize=30,
                #horizontalalignment='center')

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label='$sin2x$')
axes[1, 0].fill_between(x, fun1_x, color='b', alpha=.125)
axes[1, 0].set_title('$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])
axes[1, 0].legend(loc='upper right', frameon=True)
#axes[1, 0].text(0, 0, 'I am [1, 0]!', fontsize=30,
                #horizontalalignment='center')

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label='$sin 2x$')
axes[1, 1].set_title('$e^{-x}\sin 2x$')
axes[1, 1].fill_between(x, fun2_x, 10, color='y', alpha=.125)
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].set_xlabel('Angle (radians)')
axes[1, 1].legend(loc='upper right', frameon=True)
#axes[1, 1].text(0, 0, 'I am [1, 1]!', fontsize=30,
               # horizontalalignment='center')

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.25)
    

plt.tight_layout()
plt.show()

# How do i change the limits of the filled areas? 

```
