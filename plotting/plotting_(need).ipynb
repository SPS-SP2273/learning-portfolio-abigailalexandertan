## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
# Load matplotlib

from matplotlib import pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Start plotting!
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

Longform vs abbreviation:
- Colour = c
- Linestyle = ls
- Linewidth = lw

```python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)

#IS THE SAME AS# 

plt.plot(x, y_1, c='red', ls='dashed', lw=2)

plt.show()
```

```python
# plots can be kept really simple! Plotting functions usually
# have default values already. 

plt.plot(x, y_1)

plt.show()
```

### 1.2 Adding another plot

```python
# Another plot can be added. Another plot = y_2 in blue color

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, c='red', ls='dashed', label='Y values')
plt.plot(x, y_2, c='blue', label='Y2 values') #ADD PLOT
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

### 1.3 Yet another plot but with error bars

```python
# x and y error bars can also be added to a data line.
# the plotting command for this is errorbar()

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, c='red', ls='dashed', label='Y values')
plt.plot(x, y_2, c='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,  #TO PLOT ERROR BAR
            color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

### 1.5 Better with NumPy

```python
# It is easier to use NumPy arrays than Python lists. So we can convert
# the Python list into a NumPy array and redo the plot.

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)      # Change Python list to NumPy arrays first
np_y_1 = np.array(y_1)  # Then it will be easier to do the plt.plot step
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c='red', ls='dashed', label='Y values')
plt.plot(np_x, np_y_2, c='blue', label='Y2 values')
         
plt.show()
```

### 1.6 Adding mathematical functions

```python
# NumPy arrays allows us to generate  
# data-related mathematical functions easily!

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, c='red', ls='dashed', label='x^2')
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend()

plt.show()
```

### 1.7 We need another axis!

```python
# In order to form another y-axis that shares the same x-axis, use
# the command twinx().

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, c='red', ls='dashed', label='x^2')
plt.legend(loc='lower left')

#by placing plt.twinx() here, a new y-axis for the plot  
#that comes after plt.twin() will be created.
plt.twinx() 
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

```python
# The above plot looks choppy because there are few points only.
# To make it smooooooth, use np.linspace!

x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, c='red', ls='dashed', label='x^2')
plt.legend(loc='lower left')

#by placing plt.twinx() here, a new y-axis for the plot  
#that comes after plt.twin() will be created.
plt.twinx() 
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python

# We first save to disc if we want to use the plot in a report.
# Use the function savefig() and specify the format with the extension
#(e.g. filename.pdf) of the file name.
# Specify resolution by specifying dots-per-inch (dpi)

x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, c='red', ls='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('we-need-another-axis.png', dpi=150)

# After doing so, the image will be saved in the same folder as 
# the current notebook. 
# In order to save as a JPEG, use 'we-need-another-axis.jpeg'.
# In order to increase resolution of the image, increase dpi.
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
# First look at data with a simple plot. 

plt.style.use('fivethirtyeight') 
# this is to use a standard style template
# if you wanna use default style, change to 'default'


data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]



plt.plot(date, anomaly, alpha=0.5) # alpha is at a low value 
                                   # to soften the colour of the plot 

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
# this is to plot a horizontal line at the zero value 

plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to average from Jan 1951 - Dec 1980.)')

plt.show()
```
